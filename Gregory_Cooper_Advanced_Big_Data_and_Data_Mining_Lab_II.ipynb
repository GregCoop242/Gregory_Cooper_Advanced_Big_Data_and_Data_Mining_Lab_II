```python
import pandas as pd
from sklearn.datasets import load_wine

#show data and put in dataframe
wine = load_wine()

df = pd.DataFrame(wine.data, columns = wine.feature_names)
print(df.head())
```

```python
df.info()
```

```python
df.describe()
```

```python
df['target'] = wine.target
print(df.head())
```

```python
from sklearn.model_selection import train_test_split

# Features (X) and target (y)
X = df.drop('target', axis=1)
y = df['target']

# Split into training and testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,      # 20% test, 80% train
    random_state=42     # for reproducibility
)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier as kn
#create var with n = 1,5,11,15,21
k_1 = kn(n_neighbors=1)
k_5 = kn(n_neighbors=5)
k_11 = kn(n_neighbors=11)
k_15 = kn(n_neighbors=15)
k_21 = kn(n_neighbors=21)

k_1.fit(X_train, y_train)
k_5.fit(X_train, y_train)
k_11.fit(X_train, y_train)
k_15.fit(X_train, y_train)
k_21.fit(X_train, y_train)







```

```python
# make predictions
y_pred_1 = k_1.predict(X_test)
y_pred_5 = k_5.predict(X_test)
y_pred_11 = k_11.predict(X_test)
y_pred_15 = k_15.predict(X_test)
y_pred_21 = k_21.predict(X_test)
```

```python
#evaluate models
from sklearn.metrics import accuracy_score

accuracy_1 = accuracy_score(y_test, y_pred_1)
accuracy_5 = accuracy_score(y_test, y_pred_5)
accuracy_11 = accuracy_score(y_test, y_pred_11)
accuracy_15 = accuracy_score(y_test, y_pred_15)
accuracy_21 = accuracy_score(y_test, y_pred_21)

print(f"Accuracy with k={1}: {accuracy_1:.4f}")
print(f"Accuracy with k={5}: {accuracy_5:.4f}")
print(f"Accuracy with k={11}: {accuracy_11:.4f}")
print(f"Accuracy with k={15}: {accuracy_15:.4f}")
print(f"Accuracy with k={21}: {accuracy_21:.4f}")


```

```python
#Step 3 Implement Radius Neighbors (RNN)
from sklearn.neighbors import RadiusNeighborsClassifier as rnn

r_350 = rnn(radius=350)
r_400 = rnn(radius=400)
r_450 = rnn(radius=450)
r_500 = rnn(radius=500)
r_550 = rnn(radius=550)
r_600 = rnn(radius=600)

#fit the training data
r_350.fit(X_train,y_train)
r_400.fit(X_train,y_train)
r_450.fit(X_train,y_train)
r_500.fit(X_train,y_train)
r_550.fit(X_train,y_train)
r_600.fit(X_train,y_train)

#make predictions
r_350_pred = r_350.predict(X_test)
r_400_pred = r_400.predict(X_test)
r_450_pred = r_450.predict(X_test)
r_500_pred = r_500.predict(X_test)
r_550_pred = r_550.predict(X_test)
r_600_pred = r_600.predict(X_test)

#Get accuracy - evaluate models
accuracy_350 = accuracy_score(y_test,r_350_pred)
accuracy_400 = accuracy_score(y_test,r_400_pred)
accuracy_450 = accuracy_score(y_test,r_450_pred)
accuracy_500 = accuracy_score(y_test,r_500_pred)
accuracy_550 = accuracy_score(y_test,r_550_pred)
accuracy_600 = accuracy_score(y_test,r_600_pred)

print(f"Accuracy with R={350}: {accuracy_350:.4f}")
print(f"Accuracy with R={400}: {accuracy_400:.4f}")
print(f"Accuracy with R={450}: {accuracy_450:.4f}")
print(f"Accuracy with R={500}: {accuracy_500:.4f}")
print(f"Accuracy with R={550}: {accuracy_550:.4f}")
print(f"Accuracy with R={600}: {accuracy_600:.4f}")


```

```python
#Step 4: Visualize and Compare Results 
import matplotlib.pyplot as plt
knn_var = [1,5,11,15,21]
rnn_var = [350,400,450,500,550,600]
knn_acc = [accuracy_1,accuracy_5,accuracy_11,accuracy_15,accuracy_21]
rnn_acc = [accuracy_350,accuracy_400,accuracy_450,accuracy_500,accuracy_550,accuracy_600]
plt.xlabel("K values")
plt.ylabel("K accuracy score")
plt.plot(knn_var,knn_acc)
plt.show()
```

```python
plt.xlabel("Radius values")
plt.ylabel("Radius accuracy score")
plt.plot(rnn_var,rnn_acc)
plt.show()
```

```python
#From the Graphs, 
#K when there is a low k-value used there is a high accuracy, as intermediate k-neighbor values are used there is a downward trend
#indicating that a middle amount is not the best for this classifier however when a larger amount of k values(21) is used
#There returns a high increase in accuracy


#When the Radius Classifier is used on the data it starts off with high accuracy just a bit lower than the original accuracy for k-values.
# As additional radius' are used that accuracy significantly drops until it platueas downward and remains. This may indiciate it is not the most
#favorable classifier for this particular dataset


#K Nearest Neighbor Classifier has the highest accuracy out of both points even with its intermediate drop
```
